````python
names = ["simon", "Jesus", "Jhon"]
i = 0

while i<3:
    if i == 1:
        i+=1
        continue
    print(names[i])
    i+=1
    

    
````

````python
# WAP to calculate sum of first N natural numbers using while loop.

num = int(input("Enter a number: "))
i = 1
sum = 0
while i <= num:
    sum += i
    i+=1

print("Sum is: ", sum)
````

````python
# WAP to calculate multiple of first N natural numbers using while loop.

num = int(input("Enter a number: "))
i = 1
result = 1
while i <= num:
    result *= i
    i+=1

print("Multiple is: ", result)
````

## Python Comprehension

````python
squared = {i:i**2 for i in range(1,20) if i%2 == 0}
squared
````

````python
## Count length of words (if length > 5 )
 
Input:
```
    sent = "A single training set that has already been processed is usually split into several types of datasets in machine learning which is needed to check how well the training of the model went"
```
 
Sample OP:
```{'single': 6, 'training': 8, 'already': 7, 'processed': 9, 'usually': 7, 'several': 7, 'datasets': 8}```
 
````

````python
sent = "A single training set that has already been processed is usually split into several types of datasets in machine learning which is needed to check how well the training of the model went"
result = {words: len(words) for words in sent.split() if len(words) > 5}
result
````
